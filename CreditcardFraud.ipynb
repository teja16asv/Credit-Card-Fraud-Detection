```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading data
dataset=pd.read_csv("E:\Project2\creditcard.csv")
```

```python
dataset.head()
```

```python
# last 5 rows
dataset.tail()
```

```python
dataset.info()
```

```python
#check no. of missing values
dataset.isnull().sum()
```

```python
dataset['Class'].value_counts()
```

```python
#seperating data
correct=dataset[dataset.Class == 0]
fraud=dataset[dataset.Class == 1]
```

```python
print(correct.shape)
```

```python
# statistical measures
correct.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing values
dataset.groupby('Class').mean()
```

```python
#creating sample dataset
#N=492 MEANS random 492 transactions
correct_sample=correct.sample(n=492)
```

```python
#combining two data frames
# axis=0 means row wise (or) column wise
new_dataset=pd.concat([correct_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting into features and targets
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
# split into train and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
# training model
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy= accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy on training data : ", training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy score on test data: ", test_data_accuracy)
```
